```python
# @title CALCULOS
# =============================================
# üßÆ BLOQUE 1. C√ÅLCULOS DE SUELDO
# =============================================
# === 3. VARIABLES DE ENTRADA ===
sueldo_base = float(sueldo_base)
utilidades = float(Utilidades)
bono = float(Bono)
transporte = float(Transporte)

fecha_inicio_dt = datetime.strptime(fecha_inicio, "%d/%m/%Y")
fecha_fin_dt = datetime.strptime(fecha_fin, "%d/%m/%Y")

# === 4. CALCULOS DE SUELDO ===
fondo_reserva = round(sueldo_base * 8.33 / 100, 2) if str(posee_fondo_reserva).lower() == "si" else 0
ingreso_sueldo = sueldo_base + fondo_reserva
porcentaje_iess = 9.45
aporte_iees = round(sueldo_base * porcentaje_iess / 100, 2)
prestamo_quirografario = valor_prestamo_quirografario if str(posee_prestamo_quirografario).lower() == "si" else 0

# Totales incluyendo nuevas variables
total_ingresos = ingreso_sueldo + utilidades + bono
total_deducciones = aporte_iees + prestamo_quirografario + transporte
total_pagar_calc = round(total_ingresos - total_deducciones, 2)

# Convertir a string
sueldo_base_str = f"{sueldo_base:.2f}"
ingreso_sueldo_str = f"{ingreso_sueldo:.2f}"
aporte_iees_str = f"{aporte_iees:.2f}"
prestamo_quirografario_str = f"{prestamo_quirografario:.2f}"
total_ingresos_str = f"{total_ingresos:.2f}"
total_deducciones_str = f"{total_deducciones:.2f}"
total_pagar_str = f"{total_pagar_calc:.2f}"
fondo_reserva_str = f"{fondo_reserva:.2f}"
utilidades_str = f"{utilidades:.2f}"
bono_str = f"{bono:.2f}"
transporte_str = f"{transporte:.2f}"

# === 5. FUNCIONES AUXILIARES ===
def dias_trabajados(inicio_mes, fin_mes):
    inicio_dt = datetime.strptime(inicio_mes, "%d/%m/%Y")
    fin_dt = datetime.strptime(fin_mes, "%d/%m/%Y")
    return (fin_dt - inicio_dt).days + 1

def meses_en_rango(fecha_inicio, fecha_fin):
    fechas = []
    current = fecha_inicio.replace(day=1)
    while current <= fecha_fin:
        inicio_mes = current
        fin_mes = (current + relativedelta(months=1)) - timedelta(days=1)
        if fin_mes > fecha_fin:
            fin_mes = fecha_fin
        fechas.append((inicio_mes.strftime("%d/%m/%Y"), fin_mes.strftime("%d/%m/%Y")))
        current += relativedelta(months=1)
    return fechas

def reemplazar_en_runs(parrafos, reemplazos):
    for parrafo in parrafos:
        for marcador, valor in reemplazos.items():
            while marcador in parrafo.text:
                acumulado = ""
                run_indices = []
                for i, run in enumerate(parrafo.runs):
                    acumulado += run.text
                    run_indices.append(i)
                    if marcador in acumulado:
                        resto = acumulado.replace(marcador, str(valor))
                        for j in run_indices:
                            parrafo.runs[j].text = ""
                        parrafo.runs[run_indices[0]].text = resto
                        break
```

```python
# @title MANIPULAION DOCUMENTO
# =============================================
# üß± BLOQUE 3. FUNCIONES DOCX
# =============================================
def poner_fondo_gris(cell):
    tc = cell._tc
    tcPr = tc.get_or_add_tcPr()
    shd = OxmlElement('w:shd')
    shd.set(qn('w:fill'), "D9D9D9")
    tcPr.append(shd)

def agregar_tablas_separadas(doc, reemplazos):
    section = doc.sections[0]
    section.top_margin = Cm(1)

    # === T√çTULO PRINCIPAL ===
    titulo = doc.add_paragraph("ROL DE PAGOS INDIVIDUAL UNIDAD EDUCATIVA SAN FRANCISCO DE ASIS")
    titulo.alignment = WD_ALIGN_PARAGRAPH.CENTER

    # === INFORMACI√ìN EMPLEADO ===
    info_texto = (
        f"Empleado: {Nombre_empleado} ‚Äì {cargo}\n"
        f"Departamento: {Departamento}\n"
        f"Fecha Desde: {reemplazos['EDIT_06']}           Fecha Hasta: {reemplazos['EDIT_07']}                 D√≠as Trabajados: {reemplazos['EDIT_08']}"
    )
    doc.add_paragraph(info_texto)

    # === TABLA PRINCIPAL: INGRESOS / DEDUCCIONES ===
    tabla1 = doc.add_table(rows=1, cols=2)
    tabla1.style = 'Table Grid'

    # Encabezados
    hdr1 = tabla1.rows[0].cells
    hdr1[0].text = "INGRESOS"
    hdr1[1].text = "DEDUCCIONES"
    poner_fondo_gris(hdr1[0])
    poner_fondo_gris(hdr1[1])

    # === LISTAS DIN√ÅMICAS DE INGRESOS Y DEDUCCIONES ===
    ingresos = []
    deducciones = []

    # INGRESOS
    ingresos.append(f"Sueldo Base: ${sueldo_base:.2f}")
    if fondo_reserva > 0:
        ingresos.append(f"Fondo de Reserva: ${fondo_reserva:.2f}")
    if utilidades > 0:
        ingresos.append(f"Utilidades: ${utilidades:.2f}")
    if bono > 0:
        ingresos.append(f"Bono: ${bono:.2f}")

    # DEDUCCIONES
    deducciones.append(f"Aporte IESS: ${aporte_iees:.2f}")
    if prestamo_quirografario > 0:
        deducciones.append(f"Pr√©stamo Quirografario IESS: ${prestamo_quirografario:.2f}")
    if transporte > 0:
        deducciones.append(f"Transporte: ${transporte:.2f}")

    # Determinar cu√°ntas filas crear (la mayor cantidad entre ingresos y deducciones)
    num_filas = max(len(ingresos), len(deducciones))

    # Crear filas sin dejar celdas vac√≠as
    for i in range(num_filas):
        fila = tabla1.add_row().cells
        fila[0].text = ingresos[i] if i < len(ingresos) else ""
        fila[1].text = deducciones[i] if i < len(deducciones) else ""

    doc.add_paragraph("\n")

    # === TABLA DE TOTALES ===
    tabla2 = doc.add_table(rows=2, cols=6)
    tabla2.style = 'Table Grid'

    # Ajustar ancho de columnas
    ancho_columnas = [Cm(3)] * 6
    for col, ancho in zip(tabla2.columns, ancho_columnas):
        for cell in col.cells:
            cell.width = ancho

    # Encabezados de totales
    hdr2 = tabla2.rows[0].cells
    hdr2[0].text = "Total Ingreso"
    hdr2[1].text = f"${reemplazos['EDIT_02']}"
    hdr2[2].text = "Total Deducciones"
    hdr2[3].text = f"${reemplazos['EDIT_03']}"
    hdr2[4].text = "Total a Pagar"
    hdr2[5].text = f"${reemplazos['EDIT_04']}"

    poner_fondo_gris(hdr2[0])
    poner_fondo_gris(hdr2[2])
    poner_fondo_gris(hdr2[4])

    for cell in tabla2.rows[0].cells:
        for paragraph in cell.paragraphs:
            paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER
        cell.vertical_alignment = WD_ALIGN_PARAGRAPH.CENTER

    # === FILA DE FORMA DE PAGO ===
    fila_pago = tabla2.rows[1].cells
    fila_pago[0].text = "Forma de pago"
    fila_pago[1].merge(fila_pago[5])
    fila_pago[1].text = reemplazos["EDIT_0016"]

    poner_fondo_gris(fila_pago[0])
    fila_pago[0].paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.CENTER
    fila_pago[1].paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.CENTER
    for cell in fila_pago:
        cell.vertical_alignment = WD_ALIGN_PARAGRAPH.CENTER


# === NUEVA FUNCI√ìN DE FIRMA ===
def agregar_texto_firma(doc, espacio_cm=1):
    """
    Agrega espacio y l√≠neas de firma para empleado y empleador.
    """
    doc.add_paragraph("\n" * int(espacio_cm))
    doc.add_paragraph("\n_____________________________").alignment = WD_ALIGN_PARAGRAPH.CENTER
    doc.add_paragraph("RECIB√ç CONFORME").alignment = WD_ALIGN_PARAGRAPH.CENTER

```

```python
# @title ENVIAR CORREO

# === FUNCI√ìN PARA BUSCAR CORREO Y ENVIAR ROL ===
def enviar_rol_por_correo(nombre_empleado, ruta_archivo):
    """
    Busca el correo del empleado en una hoja de Google Sheets
    y env√≠a el archivo .docx al correo correspondiente.
    """
    import smtplib
    from email.mime.multipart import MIMEMultipart
    from email.mime.text import MIMEText
    from email.mime.base import MIMEBase
    from email import encoders
    import mimetypes

    # === 1. ABRIR LA HOJA DE EMPLEADOS ===
    hoja_empleados = gc.open_by_url("https://docs.google.com/spreadsheets/d/1NXI1mHnySsjBst1OMnHfkvCeeFDTI9UbGbeCc7RScwI/edit?gid=0#gid=0").sheet1
    datos = hoja_empleados.get_all_records()

    # === 2. BUSCAR EL EMPLEADO ===
    correo_destino = None
    for fila in datos:
        if str(fila.get("Nombre_empleado", "")).strip().lower() == nombre_empleado.strip().lower():
            correo_destino = fila.get("Correo") or fila.get("correo") or fila.get("EMAIL")
            break

    if not correo_destino:
        print(f"‚ö†Ô∏è No se encontr√≥ correo para el empleado: {nombre_empleado}")
        return

    # === 3. CONFIGURAR ENV√çO DE CORREO ===
    remitente = "rolalfredouesfa@gmail.com"  # ‚Üê pon aqu√≠ tu cuenta institucional o de env√≠o
    clave_app = ""#"vcvt vrww iwkf tvzy"  # ‚Üê clave de aplicaci√≥n (no la contrase√±a normal)
    asunto = f"Rol de pagos - {nombre_empleado}"
    mensaje = "S√≠rvanse a encontrar el rol adjunto."

    # === 4. CREAR EL MENSAJE ===
    msg = MIMEMultipart()
    msg["From"] = remitente
    msg["To"] = correo_destino
    msg["Subject"] = asunto
    msg.attach(MIMEText(mensaje, "plain"))

    # === 5. ADJUNTAR ARCHIVO ===
    ctype, encoding = mimetypes.guess_type(ruta_archivo)
    maintype, subtype = (ctype or "application/octet-stream").split("/", 1)
    with open(ruta_archivo, "rb") as adjunto:
        parte = MIMEBase(maintype, subtype)
        parte.set_payload(adjunto.read())
        encoders.encode_base64(parte)
        parte.add_header("Content-Disposition", f"attachment; filename={os.path.basename(ruta_archivo)}")
        msg.attach(parte)

    # === 6. ENVIAR EL CORREO ===
    try:
        with smtplib.SMTP("smtp.gmail.com", 587) as servidor:
            servidor.starttls()
            servidor.login(remitente, clave_app)
            servidor.send_message(msg)
        print(f"üì§ Rol enviado a {correo_destino}")
    except Exception as e:
        print(f"‚ùå Error enviando correo a {correo_destino}: {e}")

```
